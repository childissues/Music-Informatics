## 1. Introduction

The task of composer identification involves automatically determining the author of a musical piece given its symbolic representation (in this case, MusicXML). This is a fascinating problem both from a musical and technical standpoint.

**Musical Issues:** Composers often have unique "fingerprints" in terms of harmonic language, melodic tendencies, rhythmic patterns, and use of texture. However, these styles can overlap within the same era (e.g., Mozart and Haydn) or even change throughout a composer's career.

**Technical Issues:** Symbolic music data is high-dimensional and non-uniform. Extracting meaningful features that capture stylistic nuances—rather than just mechanical properties—is the primary challenge. Additionally, the dataset size is relatively small, which requires robust models that avoid overfitting.

## 2. Description of Methods

### Why this approach?
We chose a **feature-based classification** approach using a **Random Forest** model. 
- **Feature Selection:** Instead of raw pianorolls, which can be sparse and sensitive to temporal variations, we extracted statistical distributions of musical parameters (pitches, durations, intervals). This aggregates stylized patterns over the whole piece, making the model more robust to varying piece lengths.
- **Model Selection:** Random Forest is excellent for small datasets with tabular features. It handles non-linear relationships well and provides an intrinsic measure of feature importance.

### Parameters and Musical Relation
- **Pitch Distribution:** Captures the composer's tonal palette and register preferences.
- **Duration Distribution:** Captures rhythmic style (e.g., prevalence of short notes vs. long lyrical lines).
- **Interval Distribution:** Captures melodic "jumps" vs. stepwise motion, which is a strong stylistic marker.
- **Polyphonic Density:** Measures how "thick" or "thin" the musical texture is on average.

## 3. Evaluation

### Metrics
We evaluate our model using **Accuracy** and **F1-Score**. Accuracy provides a general sense of correctness, while F1-Score ensures we perform well across all composer classes, especially if some are underrepresented.

### Dataset and Features
The dataset contains 247 labeled MusicXML files for training and 61 unlabeled files for testing. We use 128-dimensional pitch distributions, discretized duration buckets, and melodic interval bins as our primary features.

### Training and Hyperparameters
The model was trained using 5-fold Stratified Cross-Validation to ensure reliable performance estimates. 
- **N-estimators:** 200
- **Max-depth:** 20
- **Class weights:** Balanced (to account for class imbalance)

**Cross-Validation Results:**
- Mean Accuracy: ~68.7%
- Standard Deviation: ~3.3%

## 4. Discussion and Conclusions

### What worked?
The statistical features (especially pitch and interval distributions) proved quite effective. The Random Forest model's ability to handle high-dimensional feature vectors without much tuning was a significant advantage.

### Limitations
- **Temporal Information:** Our current features are global averages. We lose the sequence information (e.g., specific chord progressions or formal structures).
- **Part Selection:** We mainly focused on the first part of the MusicXML. For complex orchestral or multi-staff pieces, information might be lost.

### Conclusion
Musical style is deeply embedded in the distribution of notes and intervals. While a 68.7% CV accuracy is promising, a more advanced approach using sequence models (like LSTMs or Transformers) or 2D CNNs on pianorolls might capture deeper structural relationships, provided enough data is available.
